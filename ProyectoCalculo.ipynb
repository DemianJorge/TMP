```python
X = np.arange(-8, 8, 0.25)
Y = np.arange(-8, 8, 0.25)
X, Y = np.meshgrid(X, Y)

X0 = 0.0; Y0 = 0.0
R0 = np.sqrt((X0 - X)**2 + (Y0- Y)**2)

X1 = 3.0; Y1 = 3.0
R1 = np.sqrt((X1 - X)**2 + (Y1- Y)**2)

X2 = 4.0; Y2 = 0.0
R2 = np.sqrt((X2 - X)**2 + (Y2- Y)**2)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')


R = np.sqrt(X**2 + Y**2)
Z = 40-R0**0.5 + 10-R1 + 10-R2


surf = ax.plot_surface(X, Y, Z, cmap=cm.Greys,
                       linewidth=0, antialiased=False,alpha=0.5)

if Z.min()<0:
    lv = np.linspace(0,Z.max(),50)
    ax.set_zlim(0, Z.max())
else:
    lv = np.linspace(Z.min(),Z.max(),50)
    ax.set_zlim(Z.min(), Z.max())


cset = ax.contour(X, Y, Z, zdir='z', offset=Z.min(), levels=lv, cmap=cm.coolwarm)
```

```python
size = len(X)
X = X.reshape(size*size,) 
Y = Y.reshape(size*size,) 
Z = Z.reshape(size*size,) 
```
